```python
predictors = ['Age', 'Chronic', 'Smoker', 'Drinker', 'Urban', 'BDI Total', 'BAI Total', 'Gender']
target = 'Immuno'
train_data, test_data, train_sln, test_sln = train_test_split(subset_nomiss[predictors], subset_nomiss[target], test_size = 0.2, random_state=41)
train_data
```

```python
import numpy as np

print(np.isinf(train_data))

train_data = np.nan_to_num(train_data)

print(np.isinf(train_data))
```

```python
from sklearn.metrics import accuracy_score

scaler = StandardScaler()
scaler.fit(train_data)
scaler.fit(test_data)
test_scaled = scaler.fit_transform(test_data)

train_scaled = scaler.fit_transform(train_data)

from sklearn import tree
from sklearn import metrics

dt = tree.DecisionTreeClassifier(max_depth = 40, random_state = 1, class_weight='balanced')

dt.fit(train_data,train_sln)

immuno_pred = dt.predict(test_data)

print("Unnormalized data results")
accuracy = accuracy_score(immuno_pred, test_sln)

# Print the accuracy score
print('Accuracy:', accuracy)

train_sln.value_counts()
test_sln.value_counts()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model2 = Sequential() 
#build your model here:
model2.add(Dense(128, activation='relu', input_dim=8))
model2.add(Dense(1, activation='sigmoid')) # binary classification 
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
model2.summary()
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Initialize the model
model = Sequential()

# Add the input layer and first hidden layer
model.add(Dense(1, activation='relu', input_dim=8))

# Add additional hidden layers
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# Add the output layer
model.add(Dense(1, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
hist2 = model.fit(ntrain_data, ntrain_sln, validation_data=(ntest_data, ntest_sln), epochs=10, batch_size=100)
```
